```python

import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
```

```python
import kagglehub
import pandas as pd
import os

# Download dataset
path = kagglehub.dataset_download("snap/amazon-fine-food-reviews")

print("Dataset Path:", path)
print("Files:", os.listdir(path))

# Load Reviews.csv
csv_path = os.path.join(path, "Reviews.csv")

df = pd.read_csv(csv_path, nrows=5000)

print(df.head())
print(df.shape)
```

```python
light_reviews = df[df["Text"].str.contains(r"\blight\b", case=False, na=False)]

print("Total reviews containing 'light':", len(light_reviews))

# Display a few examples to manually choose sentences
for i, review in enumerate(light_reviews["Text"].head(20)):
    print(f"\nReview {i+1}:")
    print(review)
```

```python
sentence_1 = "This olive oil is wonderful and very light on the stomach."
sentence_2 = "The packaging was broken and let too much light inside the box."
sentence_3 = "I prefer this brand because it is a light and healthy snack."

sentences = [sentence_1, sentence_2, sentence_3]
```

```python
vectorizer = CountVectorizer()

doc_term_matrix = vectorizer.fit_transform(sentences)

dense_matrix = doc_term_matrix.toarray()

print("\nDense Matrix:")
print(dense_matrix)

print("\nVocabulary:")
print(vectorizer.get_feature_names_out())
```

```python
vectorizer = CountVectorizer()

doc_term_matrix = vectorizer.fit_transform(sentences)

dense_matrix = doc_term_matrix.toarray()

print("\nDense Matrix:")
print(dense_matrix)

print("\nVocabulary:")
print(vectorizer.get_feature_names_out())
```

```python
vec_1 = dense_matrix[0]
vec_2 = dense_matrix[1]
vec_3 = dense_matrix[2]

print("\nVector 1:", vec_1)
print("Vector 2:", vec_2)
print("Vector 3:", vec_3)
```

```python
similarity_1_2 = np.dot(vec_1, vec_2)

similarity_1_3 = np.dot(vec_1, vec_3)
```

```python
print("\n==============================")
print("Similarity Scores")
print("==============================")

print("Sentence 1 vs Sentence 2 :", similarity_1_2)
print("Sentence 1 vs Sentence 3 :", similarity_1_3)
```
